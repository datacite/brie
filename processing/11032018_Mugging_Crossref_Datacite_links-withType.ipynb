```r
get_type<-function(doi){
    doi_name<-str_extract(doi,'(10\\.\\d{4,5}\\/.+)\\z')
#     print(doi_name)
    dc_info<-GET(paste0('https://api.datacite.org/works/',doi_name))
    if(!is.null(dc_info$erros)){
        return("xxxxxx")
    }
    parsed <- content(dc_info, "parsed")
    if(!is.null(parsed$data$attributes$resource_type_id)){
            r <- (parsed$data$attributes$resource_type_id)
    }else{
        r<- ""
    }

    return(r)   
}
get_type('https://api.datacite.org/works/10.13140/rg.2.2.34715.57121')
```

### Adding subjects to crossref subjects

```r
get_subject<-function(doi){
    doi_name<-str_extract(doi,'(10\\.\\d{4,5}\\/.+)\\z')
#     print(doi_name)
    dc_info<-GET(paste0('https://api.crossref.org/works/',doi_name))
    if(!is.null(dc_info$erros)){
        return("xxxxxx")
    }
    parsed <- content(dc_info, "parsed")
    if(!is.null(head(parsed$message$subject,1))){
            r <- head(parsed$message$subject,1)
    }else{
        r<- ""
    }

    return(r)   
}
get_subject('https://api.crossref.org/works/10.1016/s0305-9006(98)00020-8')
```

```r
extract_subj_doi<-function(data){
    data <- data %>%  
        mutate(doi = str_extract(subj_id,'(10\\.\\d{4,5}\\/.+)\\z'))
#     print(nrow(data))
    return(data)
}

get_listOf_unique_subj_DOIS<-function(datas){
    data<-extract_subj_doi(datas)
    print(nrow(data))
    ss<-as.data.frame(unique(data$doi))
    colnames(ss)[1] <- "subj_id"
    print(nrow(ss))
    return(ss)
#     dc_dois<-paste(ss$`unique(data$doi)`,collapse=",") 
#     return(dc_dois)
}

uniquelist<-get_listOf_unique_subj_DOIS(data)
```

```r
get_subjects_for_dois<-function(data){
    data <- data %>% 
        rowwise() %>% 
        mutate(subject = (get_subject(subj_id)))
    print(nrow(data))
    return(data)
}
data_w_subject<-get_subjects_for_dois((uniquelist))
data_w_subject
```
